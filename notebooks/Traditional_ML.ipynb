```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import RandomForestClassifier
from collections import Counter
```

```python
df = pd.read_csv('../data/main_derived.csv')
df.head()
```

```python
# Reorder columns
df = df[['userId', 'distance', 'duration', 'calories', 'heart_rate', 'gender', 'sport']]

# Encoder to encode categorical values
encoder = OrdinalEncoder()
df = encoder.fit_transform(df)
```

```python
# Split into X and Y
X, y = df[:, :-1], df[:, -1]
```

```python
# Find best params for RandomForest
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [20, 50, None],
    'n_estimators': [10, 20, 50],
    'n_jobs': [8]
}


clf = GridSearchCV(RandomForestClassifier(), params, scoring='accuracy')
```

```python
# Fit model
clf.fit(X, y)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
# Count y frequency
y_freq = Counter(y)
```

```python
# Calculate the distribution of labels
sports_lookup = {}
for i, sport in enumerate(encoder.categories_[-1]):
    sports_lookup[i] = sport

total_y = sum(list(y_freq.values()))
for key, value in y_freq.items():
    print(f'{sports_lookup[key]:30}: {round(value / total_y * 100, 2):2}%')
```

The most popular type of activities are bike, run, outdoor bike (bike transport) and mountain bike. That's like 90% of the labels.
